```python
import pandas as pd
```

```python
old_dataset = pd.read_csv('pnl_dataset/canada_cleaned.csv')
```

```python
old_dataset = old_dataset[old_dataset.year >= 2017]
```

```python
#old_dataset['High-technology exports (% of manufactured exports)'].dropna().pct_change().median()
```

```python
old_dataset['High-technology exports (% of manufactured exports)'].dropna()
```

```python
def get_mock_value(df):
    for col in df.columns:
        pct_increase = df[col].dropna().pct_change().median()
        for i in df.index[1:]:
            if (pd.isnull(pct_increase) == False) and (pd.isnull(df.loc[i, col])):
                df.loc[i, col] = pct_increase *  df.loc[i - 1, col]
    return df
        
```

```python
df = get_mock_value(old_dataset)
```

```python
def generate_future_data(start_year, end_year, df_last_5_year):
    start_index = max(df.index) + 1
    for i in range(0, end_year - start_year):
        df_last_5_year.loc[i + start_index, 'year'] = start_year + i
        df_last_5_year = get_mock_value(df_last_5_year)
    return df_last_5_year
```

```python
final_df = generate_future_data(2022,2026, df)
```

```python
final_df = final_df[final_df.year >= 2022]
```

```python
final_df.to_csv('pnl_dataset/features_to_pridict.csv')
```
